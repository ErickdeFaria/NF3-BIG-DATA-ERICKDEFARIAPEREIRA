### ***NUMERO 2***

```python
#### NUMERO 1 #####

import pandas as pd


dtf = pd.read_csv('https://raw.githubusercontent.com/ErickdeFaria/trabalhobigdata/main/cursos-prouni.csv',encoding='latin', sep =';')


dtf
```

**LETRA A**

```python
#### NUMERO 2 LETRA A ####

import numpy as np


colunas_notas = ['nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla', 'nota_parcial_cotas',]


dtf[colunas_notas] = dtf[colunas_notas].fillna(0.0)

dtf
```

**LETRA B**

```python
#### NUMERO 2 LETRA B ####

grau = dtf.groupby('grau')

for grau, dds in grau:
  print('Grau:', grau)
  print(dds)
  print('\n')
```

**LETRA C**

```python
#### NUMERO 2 LETRA C ####


crs = ['MatemÃ¡tica','Medicina','Pedagogia']

flt = dtf.loc[dtf['curso_busca'].isin(crs)]

flt
```

**LETRA D**

```python
#### NUMERO 2 LETRA D ####

nota_media = dtf.groupby('uf_busca')[['nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla', 'nota_parcial_cotas']].mean(numeric_only=True)

print(nota_media)

```

**LETRA E**

```python
#### NUMERO 2 LETRA E ####


crs = ['TecnolÃ³gico']
flt = dtf.loc[dtf['grau'].isin(crs)]

flt
```

**LETRA F**

```python
#### NUMERO 2 LETRA F ####


dtf = dtf.drop("cidade_filtro", axis = 1)

dtf
```

**LETRA G**


```python
#### NUMERO 2 LETRA G ####


dtf['mensalidade'] = dtf['mensalidade'].astype(str)

dtf['mensalidade'] = dtf['mensalidade'].str.replace ('[^\d.]', '', regex=True)

dtf['mensalidade'] = pd.to_numeric(dtf['mensalidade'])


media = dtf[dtf['curso_busca'] == 'Medicina']['mensalidade'].mean()


media
```

**LETRA H**

```python
#### NUMERO 2 LETRA H ####


import numpy as np

media_integral = ['nota_integral_ampla', 'nota_integral_cotas']

flt = dtf['turno'] == 'Integral'


dtf[media_integral] = dtf[media_integral].apply(lambda x: pd.to_numeric(x, errors='coerce'))


media_ampla = dtf[flt]['nota_integral_ampla'].mean()

media_cotas = dtf[flt]['nota_integral_cotas'].mean()


print("MÃ©dia amplas: ", media_ampla)

print("MÃ©dia cotas: ", media_cotas)
```

**LETRA I**


```python
#### NUMERO 2 LETRA I ####


filtro = dtf['grau'] == 'Bacharelado'

tipos_integral = ['nota_integral_ampla','nota_integral_cotas']

integral_bacharel = dtf[filtro][tipos_integral]


estatistica = integral_bacharel.describe()


estatistica
```

**LETRA J**

```python
#### NUMERO 2 LETRA J ####


import matplotlib.pyplot as plt

dtf['nota_integral_cotas'] = pd.to_numeric (dtf['nota_integral_cotas'], errors='coerce')

integral = dtf.groupby('grau') ['nota_integral_cotas'].mean()

integral.plot.bar()


plt.xlabel('Grau dos Cursos')

plt.ylabel('Notas de Corte')

plt.title('ComparaÃ§Ã£o das Notas de Corte por Grau')

plt.show()
```
